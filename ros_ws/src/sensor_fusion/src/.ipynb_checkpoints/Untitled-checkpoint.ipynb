```c++
string nameGenerate(const string& suffix, const int length, int frame) {
    string file_name = std::to_string(frame);
    int cur_length = length-file_name.size();
    for (int a = 0; a < cur_length; a++)
        file_name = "0"+file_name;
    file_name += ".";
    file_name += suffix;
    return file_name;
};
```

```c++
void read_lab(const int frame, vector<Box2d>& boxes2d, vector<Box3d>& boxes3d, vector<float64> alphas) {
    string lab_path = lab_dir + lab_dir + nameGenerate("txt", NAME_LENGTH, frame);
    std::ifstream lab_file(lab_path.c_str(), std::ifstream::in);
    if(!lab_file.is_open()) RCLCPP_INFO (this->get_logger(), "Label File doesn't exist.");
    else {
        string line;
        string type;
        while(getline(input_file, line)) {
            std::istringstream iss(line);
            iss >> type;
            if (type == "Car" || type == "Van" || type == "Truck") {
                char tmp[600] = {0};
                for(size_t i = 0; i < line.length(); i++) tmp[i] = line[i];
                Box2d bbox2d;
                Box3d bbox3d;
                bbox2d.obj_class = type;
                sscanf(tmp, "%s %f %f %f %f %f %f %f %f %f %f %f %f %f %f", 
                    &bbox3d.obj_class, &truncated, &occluded, &alpha, 
                    &bbox2d.xmin, &bbox2d.ymin, &bbox2d.xmax, &bbox2d.xmax,
                    &bbox3d.height, &bbox3d.width, &bbox3d.length, 
                    &bbox3d.pos.x, &bbox3d.pos.y, &bbox3d.pos.z, &bbox3d.heading);
                boxes2d.push_back(bbox2d);
                boxes3d.push_back(bbox3d);
            }
        }
    }
    lab_file.close();
}
```
